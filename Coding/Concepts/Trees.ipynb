```python
"""
A collection of tree data structures in python

Sample tree:

l = {
        0: {1: 1, 2: 10},
        1: {1: 9, 2: 20},
        10: {1: 15, 2: 18},
        9: {1: 11, 2: 19},
        20: {1: 13, 2: 25},
        15: {1: 39, 2: 83},
        19: {1: 3, 2: 23},
        3: {1: 21, 2: 73 },
        73: {1: 5, 2: 105}
    }

pre = [0, 1, 9, 11, 19, 3, 21, 73, 5, 105, 23, 20, 13, 25, 10, 15, 39, 83, 18] 


@author: Avneesh Yadav
"""



class BinaryTree(object):
    def __init__(self, data, left=None, right=None):
        self.data = data # type node
        self.left = left
        self.right = right

def create_bt(d):
    n_bt = {}
    for k, v in d.iteritems():
        n_bt[k] = BinaryTree(k, None, None)
    for k, v in d.iteritems():
        if k in n_bt:
            l = n_bt.get(v[1], BinaryTree(v[1]))
            r = n_bt.get(v[2], BinaryTree(v[2]))
            n_bt[k].left = l
            n_bt[k].right = r
    return n_bt


def get_pre_rec(root, res=None):
    if res is None:
        res = []
    if root is None:
        return []
    res.append(root.data)
    if getattr(root, 'left', None) is not None:
        res = get_pre_rec(root.left, res)
    if getattr(root, 'right', None) is not None:
        res = get_pre_rec(root.right, res)
    return res

def get_pre_iter(root):
    if root is None:
        return []
    node_stack = [root]
    pre_tree = []
    while len(node_stack):
        top = node_stack.pop()
        pre_tree.append(top.data)
        if getattr(top, 'right', None) is not None:
            node_stack.append(top.right)
        if getattr(top, 'left', None) is not None:
            node_stack.append(top.left)
    return pre_tree

def get_in_rec(root, res=None):
    if res is None:
        res = []
    if root is None:
        return []
    if root.left is None and root.right is None:
        res.append(root.data)
        return []
    if root.left is not None:
        get_in_rec(root.left, res)
    res.append(root.data)
    if root.right is not None:
        get_in_rec(root.right, res)
    return res 
    

def get_post(self):
    pass

def get_pre(self):
    pass

def get_in(self):
    pass

def get_level(self):
    pass

def get_height(self):
    pass
```
