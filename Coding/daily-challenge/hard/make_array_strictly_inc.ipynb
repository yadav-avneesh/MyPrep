```python
"""
Given two integer arrays arr1 and arr2, return the minimum number of operations (possibly zero) needed to make arr1 strictly increasing.

In one operation, you can choose two indices 0 <= i < arr1.length and 0 <= j < arr2.length and do the assignment arr1[i] = arr2[j].

If there is no way to make arr1 strictly increasing, return -1.

 

Example 1:

Input: arr1 = [1,5,3,6,7], arr2 = [1,3,2,4]
Output: 1
Explanation: Replace 5 with 2, then arr1 = [1, 2, 3, 6, 7].
Example 2:

Input: arr1 = [1,5,3,6,7], arr2 = [4,3,1]
Output: 2
Explanation: Replace 5 with 3 and then replace 3 with 4. arr1 = [1, 3, 4, 6, 7].
Example 3:

Input: arr1 = [1,5,3,6,7], arr2 = [1,6,3,3]
Output: -1
Explanation: You can't make arr1 strictly increasing.
 

Constraints:

1 <= arr1.length, arr2.length <= 2000
0 <= arr1[i], arr2[i] <= 10^9

TIP:
    DP>>
        make array increasing as we go by replacement;
        1. replace when curr < prev with value just greater than prev
            cost = 1 + replace(i+1, prev=arr2[just_greater_than_prev])
        2. replace when curr > prev with value just greater than prev
            cost = min(replace(i+1, prev=arr1[i]), 1 + replace(i+1, prev=arr2[just_greater_than_prev]))
        3. break when we reach last
"""

from typing import List
from bisect import bisect_right

class Solution:
    def makeArrayIncreasing(self, arr1: List[int], arr2: List[int]) -> int:
        memo = {}
        arr2 = sorted(set(arr2))

        def min_replace(i, prev):
            if i == len(arr1):
                return 0
            if (i, prev) in memo:
                return memo[(i, prev)]
            
            replace = float('inf')

            if arr1[i] > prev:
                replace = min_replace(i+1, arr1[i])
            
            idx = bisect_right(arr2, prev)
            if idx < len(arr2):
                replace = min(replace, 1 + min_replace(i+1, arr2[idx]))
            
            memo[(i, prev)] = replace
            return memo[(i, prev)]
        
        minr = min_replace(0, -1)
        return minr if minr < float('inf') else -1
```
