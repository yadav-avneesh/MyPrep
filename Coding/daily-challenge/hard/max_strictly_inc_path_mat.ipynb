```python
"""
Given a 1-indexed m x n integer matrix mat, you can select any cell 
in the matrix as your starting cell. 

From the starting cell, you can move to any other cell in the same row 
or column, but only if the value of the destination cell is strictly 
greater than the value of the current cell. You can repeat this process 
as many times as possible, moving from cell to cell until you can no 
longer make any moves.

Your task is to find the maximum number of cells that you can visit 
in the matrix by starting from some cell.

Return an integer denoting the maximum number of cells that can be 
visited.

 

Example 1:
    Input: mat = [[3,1],[3,4]]
    Output: 2
    Explanation: 
        The image shows how we can visit 2 cells starting from row 1, column 2. 
        It can be shown that we cannot visit more than 2 cells no matter where 
        we start from, so the answer is 2. 


Example 2:
    Input: mat = [[1,1],[1,1]]
    Output: 1
    Explanation: 
        Since the cells must be strictly increasing, we can only 
        visit one cell in this example.


Example 3:
    Input: mat = [[3,1,6],[-9,5,7]]
    Output: 4
    Explanation:
        The image above shows how we can visit 4 cells starting from row 2, 
        column 1. It can be shown that we cannot visit more than 4 
        cells no matter where we start from, so the answer is 4. 
 

Constraints:
    m == mat.length 
    n == mat[i].length 
    1 <= m, n <= 105
    1 <= m * n <= 105
    -10^5 <= mat[i][j] <= 10^5

TIP:
    1. Traditional DP, DFS+memoization TLEs
    2. Improved version is to follow max path in reverse;
"""
from typing import List

# Improved !!
from collections import defaultdict
class Solution:
    def maxIncreasingCells(self, mat: List[List[int]]) -> int:
        R, C = len(mat), len(mat[0])
        rmax = defaultdict(int)
        cmax = defaultdict(int)
        valmax = defaultdict(list)
        for i in range(R):
            for j in range(C):
                valmax[mat[i][j]].append([i, j])
        valmax = sorted([(k, v) for k, v in valmax.items()], reverse=True)
        for val, indices in valmax:
            temp = []
            for (x, y) in indices:
                xy_mx = max(rmax[x], cmax[y]) + 1
                temp.append([x, y, xy_mx])
            for (x, y, xy_mx) in temp:
                rmax[x] = max(rmax[x], xy_mx)
                cmax[y] = max(cmax[y], xy_mx)
        return max([*rmax.values(), *cmax.values()])


# TLE => DFS + Memoization;
class Solution:
    def maxIncreasingCells(self, mat: List[List[int]]) -> int:
        R, C = len(mat), len(mat[0])
        memo = {}
        def traverse(x, y):
            if (x, y) in memo:
                return memo[(x, y)]
            
            curr = mat[x][y]
            path = 0
            for i in range(R):
                if i == x or mat[i][y] >= curr:
                    continue
                path = max(path, traverse(i, y))
            for j in range(C):
                if j == y or mat[x][j] >= curr:
                    continue
                path = max(path, traverse(x, j))
            memo[(x, y)] = 1 + path
            return memo[(x, y)]
        
        pathlen = 1
        for i in range(R):
            for j in range(C):
                pathlen = max(pathlen, traverse(i, j))
        
        return pathlen
```
