```python
"""
In a project, you have a list of required skills req_skills, and a list of people. 
The ith person people[i] contains a list of skills that the person has.

Consider a sufficient team: a set of people such that for every required skill in 
req_skills, there is at least one person in the team who has that skill. We can 
represent these teams by the index of each person.

For example, team = [0, 1, 3] represents the people with skills 
people[0], people[1], and people[3].
Return any sufficient team of the smallest possible size, represented by the 
index of each person. You may return the answer in any order.

It is guaranteed an answer exists.

 

Example 1:

Input: req_skills = ["java","nodejs","reactjs"], people = [["java"],["nodejs"],["nodejs","reactjs"]]
Output: [0,2]
Example 2:

Input: 
    req_skills = ["algorithms","math","java","reactjs","csharp","aws"], 
    people = [
                ["algorithms","math","java"],
                ["algorithms","math","reactjs"],
                ["java","csharp","aws"],
                ["reactjs","csharp"],
                ["csharp","math"],
                ["aws","java"]
            ]
Output: [1,2]
 

Constraints:
    1 <= req_skills.length <= 16
    1 <= req_skills[i].length <= 16
    req_skills[i] consists of lowercase English letters.
    All the strings of req_skills are unique.
    1 <= people.length <= 60
    0 <= people[i].length <= 16
    1 <= people[i][j].length <= 16
    people[i][j] consists of lowercase English letters.
    All the strings of people[i] are unique.
    Every skill in people[i] is a skill in req_skills.
    It is guaranteed a sufficient team exists.
"""

from typing import List
from functools import cache

class Solution:
    def smallestSufficientTeam(self, req_skills: List[str], people: List[List[str]]) -> List[int]:
        skill_id = {skill: idx for idx, skill in enumerate(req_skills)}

        skills = [0] * len(people)
        for i in range(len(people)):
            for skill in people[i]:
                skills[i] |= 1 << skill_id[skill]

        max_people  = list(range(len(skills)))
        skills_left = (1 << len(skill_id)) - 1

        @cache
        def min_group(idx, skills_left):
            if skills_left == 0:
                return []
            if idx == len(skills):
                return max_people
            skip = min_group(idx+1, skills_left)
            take = [idx, *min_group(idx+1, skills_left & ~skills[idx])]
            return min(skip, take, key=len)

        return min_group(0, skills_left)

class Solution:
    def smallestSufficientTeam(self, req_skills: List[str], people: List[List[str]]) -> List[int]:
        skill_id = {skill: idx for idx, skill in enumerate(req_skills)}
        pl = len(people)
        sl = len(req_skills)

        skills = [0] * pl
        for i in range(pl):
            for skill in people[i]:
                skills[i] |= 1 << skill_id[skill]

        skills_reqq = (1 << len(skill_id)) - 1
        skill_teams = [(1 << pl) - 1] * (1 << sl)
        skill_teams[0] = 0

        for skill_team in range(1, 1 << sl):
            for i in range(pl):
                curr_skills = skill_team & ~skills[i]
                if curr_skills != skill_team: # getting new skills on team
                    team_count = skill_teams[curr_skills] | (1 << i) # Add current people to team
                    if team_count.bit_count() < skill_teams[skill_team].bit_count():
                        skill_teams[skill_team] = team_count

        result = skill_teams[skills_reqq]
        worker = []
        for i in range(pl):
            if result >> i & 1:
                worker.append(i)

        return worker

class Solution:
    def smallestSufficientTeam(self, req_skills: List[str], people: List[List[str]]) -> List[int]:
        target = 0
        skill_id = {}

        for idx, skill in enumerate(req_skills):
            skill_id[skill] = idx
            target |= 1 << idx

        skills = [0] * len(people)
        for i in range(len(people)):
            for skill in people[i]:
                skills[i] |= 1 << skill_id[skill]

        max_people = list(range(len(skills)))

        @cache
        def min_group(idx, skills_so_far):
            if skills_so_far == target:
                return []
            if idx == len(skills):
                return max_people
            skip = min_group(idx+1, skills_so_far)
            take = [idx, *min_group(idx+1, skills_so_far | skills[idx])]
            return min(skip, take, key=len)

        return min_group(0, 0)
        


```
