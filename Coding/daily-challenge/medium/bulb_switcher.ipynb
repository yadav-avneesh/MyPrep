```python
"""
There are n bulbs that are initially off. You first turn on all the bulbs, then you turn off every second bulb.

On the third round, you toggle every third bulb (turning on if it's off or turning off if it's on). For the ith round, you toggle every i bulb. For the nth round, you only toggle the last bulb.

Return the number of bulbs that are on after n rounds.
 

Example 1:
    Input: n = 3
    Output: 1
    Explanation: At first, the three bulbs are [off, off, off].
    After the first round, the three bulbs are [on, on, on].
    After the second round, the three bulbs are [on, off, on].
    After the third round, the three bulbs are [on, off, off]. 
    So you should return 1 because there is only one bulb is on.

Example 2:
    Input: n = 0
    Output: 0

Example 3:
    Input: n = 1
    Output: 1
 

Constraints:
    0 <= n <= 109
    
TIP:

Each bulb is toggled whenever the number is one of the divisors of the bulb's number.
- For a bulb to be on, it has to be toggled an odd number of times.
- It's easy to see that numbers m generally have an even number of divisors D   because they come in pairs: {di,m/di}, however, among such divisor pairs in square numbers we have exactly one that consists of the root of the square repeated: {sqrt(m), sqrt(m)}
- The bulbs assigned to square numbers are exactly those that will stay on. The number of square numbers between [1,n] is precisely the integer square root of n, hence the solution.

isqrt(n)
"""



from math import sqrt
class Solution:
    def bulbSwitch(self, n: int) -> int:
        i = 1
        count = 0
        while i*i <= n:
            count += 1
            i += 1
        return count

from math import isqrt
class Solution:
    def bulbSwitch(self, n: int) -> int:
        return isqrt(n)
```
