```python
"""

Given a string containing just the characters '(' and ')', return the length of the longest valid (well-formed) parentheses 
substring
.

 

Example 1:

Input: s = "(()"
Output: 2
Explanation: The longest valid parentheses substring is "()".
Example 2:

Input: s = ")()())"
Output: 4
Explanation: The longest valid parentheses substring is "()()".
Example 3:

Input: s = ""
Output: 0
 

Constraints:

0 <= s.length <= 3 * 104
s[i] is '(', or ')'.

"""

# Simple processing based on previous in stack and adding up ints;
class Solution:
    def longestValidParentheses(self, s: str) -> int:
        stk = []
        longest = 0
        idx = 0
        for char in s:
            if char == '(':
                stk.append(char)
            else: # char = ')'
                if stk:
                    top = stk[-1]
                    if top == '(':
                        stk.pop()
                        stk.append(1)
                    elif top not in ('(', ')'):
                        num = stk.pop()
                        if stk and stk[-1] == '(':
                            stk.pop()
                            stk.append(num + 1)
                        else:
                            stk.append(num)
                            stk.append(')')
            while stk and isinstance(stk[-1], int):
                top = stk.pop()
                if stk and isinstance(stk[-1], int):
                    stk[-1] += top
                else:
                    stk.append(top)
                    break
        for val in stk:
            if val != '(' and val != ')':
                longest = max(val, longest)
        return longest*2
```

```python
# Keeping track of oc, cc works; However key thing is that you need to traverse twice; L -> R and R -> L; So basically any valid combination would be captured at oc == cc from either side;
class Solution:
    def longestValidParentheses(self, s: str) -> int:
        longest = 0
        chars = '(', ')'
        oc = cc = 0
        for char in s:
            if char == '(':
                oc += 1
            else:
                cc += 1
            if oc == cc:
                longest = max(longest, cc*2)
            elif cc > oc:
                oc = cc = 0
        oc = cc = 0
        for char in s[::-1]:
            if char == ')':
                oc += 1
            else:
                cc += 1
            if oc == cc:
                longest = max(longest, cc*2)
            elif cc > oc:
                oc = cc = 0
        return longest
```
