```python
"""
There is a one-dimensional garden on the x-axis. 
The garden starts at the point 0 and ends at the point n. 
(i.e The length of the garden is n).

There are n + 1 taps located at points [0, 1, ..., n] in the garden.

Given an integer n and an integer array ranges of length n + 1 
where ranges[i] (0-indexed) means the i-th tap can water the 
area [i - ranges[i], i + ranges[i]] if it was open.

Return the minimum number of taps that should be open to water 
the whole garden, If the garden cannot be watered return -1.



Example 1:
    Input: n = 5, ranges = [3,4,1,1,0,0]
    Output: 1
    Explanation: The tap at point 0 can cover the interval [-3,3]
    The tap at point 1 can cover the interval [-3,5]
    The tap at point 2 can cover the interval [1,3]
    The tap at point 3 can cover the interval [2,4]
    The tap at point 4 can cover the interval [4,4]
    The tap at point 5 can cover the interval [5,5]
    Opening Only the second tap will water the whole garden [0,5]

Example 2:
    Input: n = 3, ranges = [0,0,0,0]
    Output: -1
    Explanation: Even if you activate all the four taps you cannot water the whole garden.


Constraints:
    1 <= n <= 104
    ranges.length == n + 1
    0 <= ranges[i] <= 100

TIP:
    1. Pre-process ranges to track coverage on [0-n].
    2. Iterate L -> R and track covered k.
    3. Until k; k should be covered.
    4. X --> K-coverage --> next nbrs in K --> iterate over these to find next max coverage.
"""

from typing import List

class Solution:
    def minTaps(self, n: int, ranges: List[int]) -> int:
        coverage = [0]*(n+1)
        for i, r in enumerate(ranges):
            start, end   = max(i-r, 0), min(i+r, n)
            coverage[start] = 0 if start==end else max(coverage[start], end)
        visited, farthest, taps = 0, 0, 0
        for idx, cover in enumerate(coverage[:-1]):
            farthest = max(farthest, cover)
            if farthest >= n or farthest==idx:
                taps += 1
                break
            if idx == visited:
                taps += 1
                visited = farthest
        return taps if farthest>=n else -1


class Solution:
    def minTaps(self, n: int, ranges: List[int]) -> int:
        intervals = [(max(i-r, 0), min(i+r, n)) for i, r in enumerate(ranges)]
        cover = [0]*(n+1)
        for start, end in intervals:
            if start == end:
                cover[start] = 0
            else:
                cover[start] = max(cover[start], end)
        print(cover)
        nbr_end, mx_idx, taps = 0, 0, 0
        for idx, jump in enumerate(cover[:-1]):
            if mx_idx < idx:
                break
            mx_idx = max(mx_idx, jump)
            if mx_idx >= n:
                taps += 1
                break
            if idx == nbr_end:
                taps += 1
                nbr_end = mx_idx
        return taps if mx_idx>=n else -1

## Top solution
class Solution:
    def minTaps(self, n: int, ranges: List[int]) -> int:

        for i in range(n + 1):
            ranges[i] = [i - ranges[i], i + ranges[i]]
       
        ranges.sort(key = lambda x: x[0])
        ans = 1
        curr = prev = 0
        maxval = -1
        while prev < n + 1:
            while prev < n + 1 and ranges[prev][0] <= curr:
                maxval = max(maxval, ranges[prev][1])
                prev += 1
            if maxval >= n:
                return ans
            elif maxval == curr:
                return -1
            else:
                curr = maxval
                ans += 1
        return -1
```
