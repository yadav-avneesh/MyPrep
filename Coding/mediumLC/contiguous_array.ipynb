```python
"""
Given a binary array nums, return the maximum length of a contiguous subarray with an equal number of 0 and 1.
 

Example 1:

    Input: nums = [0,1]
    Output: 2
    Explanation: [0, 1] is the longest contiguous subarray with an equal number of 0 and 1.


Example 2:

    Input: nums = [0,1,0]
    Output: 2
    Explanation: [0, 1] (or [1, 0]) is a longest contiguous subarray with equal number of 0 and 1.
 

Constraints:

    1 <= nums.length <= 105
    nums[i] is either 0 or 1.

TIP:

# Incorrect solutions
    1. Stack trick
    2. If mergeable then merge else continue with stk
        a. mergable
            1. prev + curr; {curr>1, prev>1} or {curr+prev==1}
            2. prev>1 :: prev_1 + curr if prev_1 + curr == 1
    3. Coding solution took time -- try again;

# Correct
    Trick 
        a. in all subsequent ones --> consider 0 as -1 then if sum(i, j)=0 then equal number of 0 and 1
        b. Using above,
            1. If i=0 and we see sum as 0 for j; (0, j) --> equal 1 & 0
            2. if i=k and we see sum as x for j; (k, j) --> equal 1 & 0 because,
                0 -- k => x
                0 -- j => x
"""

## This worked for test cases
# Failed for ----[0,0,1,0,0,0,1,1]
##
class Solution:
    def findMaxLength(self, nums) -> int:
        l = len(nums)
        if l == 1:
            return 0
        stk = [nums[0]]
        for num in nums[1:]:
            stk.append(num)
            while len(stk) >= 2:
                curr = stk.pop()
                prev = stk.pop()
                if prev > 1:
                    if curr > 1:
                        stk.append(curr+prev)
                    else:
                        if stk:
                            prev_1 = stk.pop() # always 0 or 1
                            if prev_1 + curr == 1:
                                stk.append(2+prev)
                            else:
                                stk.extend([prev_1, prev, curr])
                                break
                        else:
                            stk.extend([prev, curr])
                            break
                else:
                    if curr <= 1:
                        if curr + prev == 1:
                            stk.append(2)
                        else:
                            stk.extend([prev, curr])
                            break
                    else:
                        stk.extend([prev, curr])
                        break
        maxl = max(stk)
        return 0 if maxl <= 1 else maxl
```

```python
# Correct solution
class Solution:
    def findMaxLength(self, nums) -> int:
        mnums = [-1 if num==0 else num for num in nums]
        sum_map = {}
        cum_sum = 0
        maxl = 0
        for idx, num in enumerate(mnums):
            cum_sum += num
            if cum_sum == 0:
                maxl = max(maxl, idx+1)
            else:
                if cum_sum not in sum_map:
                    sum_map[cum_sum] = idx
                else:
                    maxl = max(maxl, (idx-sum_map[cum_sum]))
        return maxl
```
