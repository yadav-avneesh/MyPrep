```python
"""
Given an array of distinct integers candidates and a target integer target, return a list of all unique combinations of candidates where the chosen numbers sum to target. You may return the combinations in any order.

The same number may be chosen from candidates an unlimited number of times. Two combinations are unique if the 
frequency
 of at least one of the chosen numbers is different.

The test cases are generated such that the number of unique combinations that sum up to target is less than 150 combinations for the given input.

 

Example 1:

Input: candidates = [2,3,6,7], target = 7
Output: [[2,2,3],[7]]
Explanation:
2 and 3 are candidates, and 2 + 2 + 3 = 7. Note that 2 can be used multiple times.
7 is a candidate, and 7 = 7.
These are the only two combinations.
Example 2:

Input: candidates = [2,3,5], target = 8
Output: [[2,2,2,2],[2,3,3],[3,5]]
Example 3:

Input: candidates = [2], target = 1
Output: []
 

Constraints:

1 <= candidates.length <= 30
2 <= candidates[i] <= 40
All elements of candidates are distinct.
1 <= target <= 40

TIP:
    Backtracking:
        Use decision tree
        Since we can't repeat, so go over next index, to not repeat again.

                                    []
                [nums[0]]   [](w/o nums[0])
            nums                nums[1:]
"""

# Inefficient --- first try
class Solution:
    def combinationSum(self, candidates, target):
        cache = {}
        def find_combination(t):
            if t in cache:
                return cache[t]
            result = []
            for num in candidates:
                if num > t:
                    continue
                if num == t:
                    result.append([num])
                else:
                    t_num = [[num, *comb] for comb in find_combination(t-num)]
                    result.extend(t_num)
            cache[t] = result
            return result
        return set([tuple(sorted(x)) for x in find_combination(target)])


# Back-tracking w/o repetition
from collections import deque
class Solution:
    def combinationSum(self, candidates, target):
        nums = sorted(candidates)
        res  = []
        idx  = 0
        def combine(idx, curr, total):
            if total == target:
                res.append(curr[::1])
                return
            if idx >= len(nums) or total > target:
                return
            curr.append(nums[idx])
            combine(idx, curr, total+nums[idx])
            curr.pop()
            combine(idx+1, curr, total)
        combine(0, [], 0)
        return res
```

```python
"""

Combination Sum : 2;

Given a collection of candidate numbers (candidates) and a target number (target), find all unique combinations in candidates where the candidate numbers sum to target.

Each number in candidates may only be used once in the combination.

Note: The solution set must not contain duplicate combinations.

 

Example 1:

Input: candidates = [10,1,2,7,6,1,5], target = 8
Output: 
[
[1,1,6],
[1,2,5],
[1,7],
[2,6]
]
Example 2:

Input: candidates = [2,5,2,1,2], target = 5
Output: 
[
[1,2,2],
[5]
]
 

Constraints:

1 <= candidates.length <= 100
1 <= candidates[i] <= 50
1 <= target <= 30

"""

class Solution:
    def combinationSum2(self, candidates: List[int], target: int):
        nums = sorted(candidates)
        result = []
        def combine(idx, curr, total):
            if total == target:
                result.append(curr)
                return
            if idx >= len(nums) or total > target:
                return
            num = nums[idx]
            curr.append(num)
            combine(idx+1, curr[::1], total+num)
            curr.pop()
            while idx < len(nums) and nums[idx] == num:
                idx += 1
            combine(idx, curr[::1], total)
        combine(0, [], 0)
        return result
```

```python
"""
Combination sum - 3;
Find all valid combinations of k numbers that sum up to n such that the following conditions are true:

Only numbers 1 through 9 are used.
Each number is used at most once.
Return a list of all possible valid combinations. The list must not contain the same combination twice, and the combinations may be returned in any order.

 

Example 1:
    Input: k = 3, n = 7
    Output: [[1,2,4]]

    Explanation:
        1 + 2 + 4 = 7
        There are no other valid combinations.

Example 2:
    Input: k = 3, n = 9
    Output: [[1,2,6],[1,3,5],[2,3,4]]

    Explanation:
        1 + 2 + 6 = 9
        1 + 3 + 5 = 9
        2 + 3 + 4 = 9
        There are no other valid combinations.

Example 3:
    Input: k = 4, n = 1
    Output: []
    
    Explanation: There are no valid combinations.
    Using 4 different numbers in the range [1,9], the smallest sum we can get is 1+2+3+4 = 10 and since 10 > 1, there are no valid combination.
 

Constraints:
    2 <= k <= 9
    1 <= n <= 60
"""

class Solution:
    def combinationSum3(self, k: int, n: int):
        result = []
        def combine(idx, curr, total):
            if total == n and len(curr)==k:
                result.append(curr)
                return
            if idx > 8 or total > n or len(curr) > k:
                return
            num = idx + 1
            curr.append(num)
            combine(idx+1, curr[::1], total+num)
            curr.pop()
            combine(idx+1, curr[::1], total)
        combine(0, [], 0)
        return result
```

```python
"""

Combination sum: 4

Given an array of distinct integers nums and a target integer target, return the number of possible combinations that add up to target.

The test cases are generated so that the answer can fit in a 32-bit integer.

 

Example 1:
    Input: nums = [1,2,3], target = 4
    Output: 7

    Explanation:
    The possible combination ways are:
    (1, 1, 1, 1)
    (1, 1, 2)
    (1, 2, 1)
    (1, 3)
    (2, 1, 1)
    (2, 2)
    (3, 1)
    Note that different sequences are counted as different combinations.


Example 2:

    Input: nums = [9], target = 3
    Output: 0
 
Constraints:
    1 <= nums.length <= 200
    1 <= nums[i] <= 1000
    All the elements of nums are unique.
    1 <= target <= 1000
 
Follow up: What if negative numbers are allowed in the given array? How does it change the problem? What limitation we need to add to the question to allow negative numbers?

TIP:
    1. Think DP
    2. Followup
        a. if negative also allowed;
        b. use map instead of list for dp;
        c. now target could be -ve as well; so dp_sum would range from (-target to target)
        d. rest works fine.
"""

# TLE
from functools import cache
class Solution:
    def combinationSum4(self, nums, target: int) -> int:
        nums = sorted(nums)
        @cache
        def combine(idx, total, comb):
            if total == target:
                return comb + 1
            if idx >= len(nums) or total > target:
                return comb
            num  = nums[idx]
            comb = combine(0, total+num, comb)
            comb = combine(idx+1, total, comb)
            return comb
        result = combine(0, 0, 0)
        return result
    

# DP Bottomup build
# Transition state --> sum value;
# T[n] = sum(T[n-num] for num in nums))
# Build bottom up fashion
class Solution:
    def combinationSum4(self, nums, target: int) -> int:
        nums = sorted(nums)

        dp = {key: 0 for key in range(target+1)}
        for num in nums:
            dp[num] = 1

        for tsum in range(target+1):
            temp = 0
            for num in nums:
                temp += dp.get(tsum-num, 0)
            dp[tsum] += temp

        return dp[target]

# Best DP solution --- mine
# Basically use as simple sum_arry to be dp
class Solution:
    def combinationSum4(self, nums, target: int) -> int:
        dp = [0]*(target+1)
        dp[0] = 1 # Takes care of num to be 1
        for tsum in range(target+1):
            for num in nums:
                if tsum - num < 0:
                    continue
                dp[tsum] += dp[tsum-num]
        return dp[-1]
```
