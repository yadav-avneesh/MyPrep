```python
"""
Given an array of distinct integers candidates and a target integer target, return a list of all unique combinations of candidates where the chosen numbers sum to target. You may return the combinations in any order.

The same number may be chosen from candidates an unlimited number of times. Two combinations are unique if the 
frequency
 of at least one of the chosen numbers is different.

The test cases are generated such that the number of unique combinations that sum up to target is less than 150 combinations for the given input.

 

Example 1:

Input: candidates = [2,3,6,7], target = 7
Output: [[2,2,3],[7]]
Explanation:
2 and 3 are candidates, and 2 + 2 + 3 = 7. Note that 2 can be used multiple times.
7 is a candidate, and 7 = 7.
These are the only two combinations.
Example 2:

Input: candidates = [2,3,5], target = 8
Output: [[2,2,2,2],[2,3,3],[3,5]]
Example 3:

Input: candidates = [2], target = 1
Output: []
 

Constraints:

1 <= candidates.length <= 30
2 <= candidates[i] <= 40
All elements of candidates are distinct.
1 <= target <= 40

TIP:
    Backtracking:
        Use decision tree
        Since we can't repeat, so go over next index, to not repeat again.

                                    []
                [nums[0]]   [](w/o nums[0])
            nums                nums[1:]
"""

# Inefficient --- first try
class Solution:
    def combinationSum(self, candidates, target):
        cache = {}
        def find_combination(t):
            if t in cache:
                return cache[t]
            result = []
            for num in candidates:
                if num > t:
                    continue
                if num == t:
                    result.append([num])
                else:
                    t_num = [[num, *comb] for comb in find_combination(t-num)]
                    result.extend(t_num)
            cache[t] = result
            return result
        return set([tuple(sorted(x)) for x in find_combination(target)])


# Back-tracking w/o repetition
from collections import deque
class Solution:
    def combinationSum(self, candidates, target):
        nums = sorted(candidates)
        res  = []
        idx  = 0
        def combine(idx, curr, total):
            if total == target:
                res.append(curr[::1])
                return
            if idx >= len(nums) or total > target:
                return
            curr.append(nums[idx])
            combine(idx, curr, total+nums[idx])
            curr.pop()
            combine(idx+1, curr, total)
        combine(0, [], 0)
        return res
```
