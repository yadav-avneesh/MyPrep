```python
"""
Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] such that i != j, i != k, and j != k, and nums[i] + nums[j] + nums[k] == 0.

Notice that the solution set must not contain duplicate triplets.

 

Example 1:

    Input: nums = [-1,0,1,2,-1,-4]
    Output: [[-1,-1,2],[-1,0,1]]

    Explanation: 
    nums[0] + nums[1] + nums[2] = (-1) + 0 + 1 = 0.
    nums[1] + nums[2] + nums[4] = 0 + 1 + (-1) = 0.
    nums[0] + nums[3] + nums[4] = (-1) + 2 + (-1) = 0.
    The distinct triplets are [-1,0,1] and [-1,-1,2].
    Notice that the order of the output and the order of the triplets does not matter.


Example 2:

    Input: nums = [0,1,1]
    Output: []

    Explanation: The only possible triplet does not sum up to 0.


Example 3:

    Input: nums = [0,0,0]
    Output: [[0,0,0]]

    Explanation: The only possible triplet sums up to 0.
 

Constraints:
    3 <= nums.length <= 3000
    -105 <= nums[i] <= 105
    
    
TIP: 
    1. Naive --> O(N^3)
    2. Improved --> sort, pick a, search b,c with target sum in O(N)
        a. O(NlogN) + O(N*N) => O(N^2)
"""

class Solution:
    def threeSum(self, nums):
        snums    = sorted(nums)
        result   = []
        prev_num = None
        for idx, num in enumerate(snums):
            if num == prev_num:
                continue
            target = 0 - num
            l, r = idx + 1, len(snums)-1
            prevl, prevr = None, None
            while l < r:
                _2sum = snums[l] + snums[r]
                if _2sum == target:
                    result.append([num, snums[l], snums[r]])
                if _2sum > target:
                    prevr = snums[r]
                    r -= 1
                    while r >= 0 and snums[r] == prevr:
                        r -= 1
                else:
                    prevl = snums[l]
                    l += 1
                    while l < len(snums) and snums[l] == prevl:
                        l += 1
            prev_num = num
        return result


# Works fine --- 
class Solution:
    def threeSum(self, nums):
        snums = sorted(nums)
        result = []
        def findsum(target, start):
            l = start + 1
            r = len(snums)-1
            res = []
            prevl, prevr = None, None
            while l < r:
                if snums[l] == prevl:
                    l += 1
                    continue
                if snums[r] == prevr:
                    r -= 1
                    continue
                _2sum = snums[l] + snums[r]
                if _2sum == target:
                    res.append([snums[l], snums[r]])
                    prevl = snums[l]
                    l += 1
                elif _2sum < target:
                    prevl = snums[l]
                    l += 1
                else:
                    prevr = snums[r]
                    r -= 1
            return res
        prev_num = None
        for idx, num in enumerate(snums):
            if num == prev_num:
                continue
            for comb in findsum(0-num, idx):
                result.append([num, *comb])
            prev_num = num
        return result
```
