```python
"""
You are given a 0-indexed array of integers nums of length n. 
You are initially positioned at nums[0].

Each element nums[i] represents the maximum length of a forward 
jump from index i. In other words, if you are at nums[i], 
you can jump to any nums[i + j] where:

0 <= j <= nums[i] and
i + j < n
Return the minimum number of jumps to reach nums[n - 1]. 
The test cases are generated such that you can reach nums[n - 1].



Example 1:
    Input: nums = [2,3,1,1,4]
    Output: 2
    Explanation: The minimum number of jumps to reach the last 
    index is 2. Jump 1 step from index 0 to 1, then 3 steps to the last index.

Example 2:
    Input: nums = [2,3,0,1,4]
    Output: 2


Constraints:
    1 <= nums.length <= 104
    0 <= nums[i] <= 1000
    It's guaranteed that you can reach nums[n - 1].
    
TIP:
    1. Concept of Dijkstra --- DP --- keep improving -- Greedy;
    2. We probably don't need to keep improving the next nbrs; 
       Just keep track of next nbr group and max we can reach through that.
    3. X ---> nbr_end ----> mx_end ----> 
        1 jump from x to nbr, if mx not end then keep jumping
       while you go through nbr_end ----> keep tracking new value for mx_end;
        1 jump from nbr_1 to new nbrs, until it doesn't have final end;
"""

class Solution:
    def jump(self, nums) -> int:
        jumps, end, mx = 0, 0, 0
        for idx, num in enumerate(nums[:-1]):
            mx = max(mx, num+idx)
            if mx >= len(nums)-1:
                jumps += 1
                break
            if end == idx:
                jumps += 1
                end = mx
        return jumps


class Solution:
    def jump(self, nums) -> int:
        jumps = [0] + [float('inf')]*(len(nums)-1)
        nl = len(nums)
        for idx, num in enumerate(nums):
            for r in range(1, min(num+1, nl)):
                jidx = min(idx+r, nl-1)
                jumps[jidx] = min(jumps[jidx], jumps[idx] + 1)
            if jumps[-1] != float('inf'):
                break
        return jumps[-1]
```
