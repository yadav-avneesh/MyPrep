```python
"""
Given an integer array nums, find a subarray that has 
the largest product, and return the product.

The test cases are generated so that the answer will fit 
in a 32-bit integer.

Example 1:
    Input: nums = [2,3,-2,4]
    Output: 6
    Explanation: [2,3] has the largest product 6.

Example 2:
    Input: nums = [-2,0,-1]
    Output: 0
    Explanation: The result cannot be 2, because [-2,-1] is not a subarray.
 

Constraints:
    1 <= nums.length <= 2 * 104
    -10 <= nums[i] <= 10
    The product of any prefix or suffix of nums is guaranteed 
    to fit in a 32-bit integer.

TIP:
    1. O(N^2): naive;
    2. DP
    3. Keep mul and check max or not; O(N)
        0. https://www.geeksforgeeks.org/maximum-product-subarray/
        a. At 0 reset to 1
        b. To track -ve to keep or not
            c. either to remove first or last
            d. hence L -> R and 1 pass from R -> L

"""

class Solution:
    def maxProduct(self, nums) -> int:
        prod = 1
        maxp = float('-inf')
        for num in nums:
            if num == 0:
                prod = 1
                maxp = max(0, maxp)
            else:
                prod *= num
                maxp = max(prod, maxp)
        prod = 1
        for num in nums[::-1]:
            if num == 0:
                prod = 1
            else:
                prod *= num
                maxp = max(prod, maxp)
        return maxp
```
