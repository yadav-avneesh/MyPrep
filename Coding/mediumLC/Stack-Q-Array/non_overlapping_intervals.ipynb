```python
## Don't think too much do simple solution first !!!
from functools import cmp_to_key
from heapq import heappush, heappop, heapify
class Solution:
    def eraseOverlapIntervals(self, intervals) -> int:
        def my_cmp(x, y):
            if x[0] > y[0]:
                return 1
            elif x[0] < y[0]:
                return -1
            else:
                if x[1] > y[1]:
                    return 1
                elif x[1] < y[1]:
                    return -1
                else:
                    return 0
        new_intervals = sorted(intervals, key=cmp_to_key(my_cmp))
        nbrs = {}
        for i in range(len(new_intervals)):
            nbrs[i] = set()
        for i in range(len(new_intervals)):
            x1, y1 = new_intervals[i]
            for j in range(i+1, len(new_intervals)):
                x2, y2 = new_intervals[j]
                if x1 <= x2 and x2 < y1:
                    nbrs[i].add(j)
                    nbrs[j].add(i)
                else:
                    break
        pq = []
        count = {}
        for node, _nbrs in nbrs.items():
            if len(_nbrs):
                pq.append((-len(_nbrs), node))
                count[node] = len(_nbrs)
        heapify(pq)
        result = 0
        while pq:
            nbrc, node = heappop(pq)
            if nbrc == 0:
                break
            if nbrc*-1 != count[node]:
                continue
            for nbr in nbrs[node]:
                count[nbr] -= 1
                nbrs[nbr].remove(node)
                heappush(pq, (-count[nbr], nbr))
            result += 1
        return result
```
