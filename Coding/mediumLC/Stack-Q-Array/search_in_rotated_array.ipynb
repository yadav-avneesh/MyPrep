```python
"""
There is an integer array nums sorted in ascending order (with distinct values).

Prior to being passed to your function, nums is possibly rotated at an unknown 
pivot index k (1 <= k < nums.length) such that the resulting array is (0-indexed)
[nums[k], nums[k+1], ..., nums[n-1], nums[0], nums[1], ..., nums[k-1]] . 
For example, [0,1,2,4,5,6,7] might be rotated at pivot index 3 and become 
[4,5,6,7,0,1,2].

Given the array nums after the possible rotation and an integer target, return 
the index of target if it is in nums, or -1 if it is not in nums.

You must write an algorithm with O(log n) runtime complexity.

Example 1:
Input: nums = [4,5,6,7,0,1,2], target = 0
Output: 4


Example 2:
Input: nums = [4,5,6,7,0,1,2], target = 3
Output: -1

Example 3:
Input: nums = [1], target = 0
Output: -1
 

Constraints:
-:: 1 <= nums.length <= 5000
-:: -104 <= nums[i] <= 104
-:: All values of nums are unique.
-:: nums is an ascending array that is possibly rotated.
-:: -104 <= target <= 104
"""

# Best --- check
# Basically we just need to see which part of array is discardable
# one of the partitioned array should be sorted !
class Solution:
    def search(self, nums, target):
        l, r = 0, len(nums)-1
        while l <= r:
            m = l + (r-l)//2
            if nums[m] == target:
                return m
            if nums[m] >= nums[l]:
                if nums[l] <= target <= nums[m]:
                    r = m - 1
                else:
                    l = m + 1
            else:
                if nums[m] < target <= nums[r]:
                    l = m + 1
                else:
                    r = m - 1
        return -1


# My solution, several bit of conditions;
# Easier could be if we check L & R values with mid;
class Solution:
    def search(self, nums, target):
        end = l = len(nums)
        start = 0
        while start < end:
            mid = start + ((end-start) // 2)
            elm = nums[mid]
            if elm == target:
                return mid
            if target == nums[start]:
                return start
            if target > elm:
                if nums[start] < elm:
                    start = mid+1
                else:
                    if target < nums[start]:
                        start = mid+1
                    else:
                        end = mid
            else:
                if nums[start] < elm:
                    if target > nums[start]:
                        end = mid
                    else:
                        start = mid+1
                else:
                    end = mid
        return -1
```
