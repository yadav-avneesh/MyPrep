```python
import pandas as pd
```

```python
d1 = ['A']*4
d2 = ['A', 'B', 'D', 'E']
d3 = ['B', 'D', 'E', 'Null']
d4 = list(range(1, 5))
rid = [111]*4
sm = ['Air']*4
col_names = ['RouteId', 'Origin', 'Warehouse', 'Destination', 'LegNum', 'Shipping']
routes = pd.DataFrame(zip(*[rid, d1, d2, d3, d4, sm]), columns=col_names)
```

```python
l1 = ['A', 'B', 'D', 'E', 'B']
l2 = ['B', 'D', 'E', None, 'D']
l3 = [1]*4+[2]
l5 = list(range(10, 60, 10))
liv_col_names = ['Warehouse', 'Destination', 'PackageDate', 'PackageCount']
livet = pd.DataFrame(zip(*[l1, l2, l3, l5]), columns=liv_col_names)
```

```python
newdfs = []
package_dates = set(livet.PackageDate.to_list())
for p_dt in package_dates:
    newdfs.append(routes.copy().assign(PackageDate=p_dt))
x1 = pd.concat(newdfs)
x1.merge(livet, how='left')
```
